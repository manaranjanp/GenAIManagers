### Initializing the OpenAI Chat Model

````python
from langchain_openai import ChatOpenAI
from langchain.chains import create_extraction_chain, create_extraction_chain_pydantic
from langchain.prompts import ChatPromptTemplate
````

````python
from langchain.chains.conversation.memory import ConversationBufferWindowMemory
from langchain.chains import RetrievalQA

# chat completion llm
llm = ChatOpenAI(
    model_name='gpt-4o-2024-11-20',
    temperature=0.0
)
````

### Configuring the Retrieval QA Chain

https://json-schema.org/understanding-json-schema/

````python
schema = {
    "properties": {
        "model_name": {"type": "string"},
        "price": {"type": "string"},
        "range_kms": {"type": "integer"}
    },
    "required": ["model_name", "price", "range_kms"],
}
````

### Configurating Conversational Memory

````python
chain = create_extraction_chain(schema, llm)
````

````python
print(chain.prompt.messages[0])
````

````python
print(chain.prompt.messages[0].prompt.template)
````

````python
chain.llm_kwargs
````

## Asking Questions

````python
response = chain.invoke(final_text)
````

````python
print(response['text'])
````

### To buy or to wait

````python
import pandas as pd
import json
````

````python
ev_df = pd.read_json(json.dumps(response['text']))
````

````python
ev_df
````

````python
ev_df['price_lakhs'] = ev_df.price.map(lambda x: (float(x.split(" ")[1])*100
                                                if 'crore' in (x.split(" ")[2].lower())
                                                else float(x.split(" ")[1])))
````

````python
ev_df['make'] = ev_df.model_name.map(lambda x: x.split()[0])
````

## Using Pydantic

https://docs.pydantic.dev/latest/

```

        "range_kms": {"type": "integer"},

    },

```

````python
from typing import Optional, List
from pydantic import BaseModel, Field
from langchain.chains import create_extraction_chain_pydantic
````

````python
from pydantic import BaseModel, Field, validator

# Pydantic data class
class Properties(BaseModel):
    car_model: str = Field(description="The name of the EV model")
    price: str = Field(description="Price of the model in lakhs (Indian Rupees)")
    range_kms: Optional[int] = Field(description="Kilometers it would travel aftera single full charge")
````

````python
chain = create_extraction_chain_pydantic(pydantic_schema=Properties,
                                         llm=llm)
````

````python
pprint(print(chain.prompt.messages[0].prompt.template))
````

````python
chain.llm_kwargs
````

````python
response = chain.invoke(final_text)
````

````python
print(response['text'])
````
